## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Notebook Info

- Fixing the counties that do not have county_FIPS in the eeoc dataset.

## CSV

```python
eeo = pd.read_csv('datasets/EEO Cleaned Data V4.csv', low_memory=False)
```

```python
counties = pd.read_csv('US Zips/uszips_all.csv')
```

```python
eeo.info()
```

```python
# 1 - Getting Missing Counties
```

```python
eeo.sort_values(by='County_FIPS')
```

```python
numeric_fips = eeo[eeo['County_FIPS'].str.isnumeric()]
numerc_fips_list = numeric_fips['County_FIPS'].tolist()
```

```python
numeric_fips
```

```python
string_fips = eeo.query("County_FIPS not in @numerc_fips_list")
string_fips

# counties that do not have a FIPS
```

```python
string_fips.to_csv("EEOC_Missing_FIPS.csv", index=False)
```

```python
print(f"There are {string_fips['County_FIPS'].nunique()} counties with missing county codes in EEOC.")
```

```python
# 2 - Getting FIPS from US ZIPS for missing EEOC 
```

```python
counties = counties.rename(columns={"county_fips":"County_FIPS", "county_name":"County", "state_id":"State"})
counties
```

```python
counties = counties.drop(['zip', 'state_name', 'population', 'lat', 'lng'], axis='columns')
```

```python
counties_clean = counties.drop_duplicates(['County', 'State'], keep='first')
counties_clean.sort_values(by='County_FIPS')
```

```python
string_fips_clean = string_fips.drop(['County_FIPS'], axis='columns')
string_fips_clean
```

```python
# Merge string_fips (eeo counties without FIPS) and counties_clean (dataset with FIPS information)

merged_missing = string_fips_clean.merge(right=counties_clean,
                  indicator=True, on=['County', 'State'], how='left')
merged_missing
```

```python
merged_both = merged_missing.query("_merge == 'both'")
merged_both
# These are the counties with county_fips that we were able to find that. They were originally missing from EEOC"
```

```python
merged_left = merged_missing.query("_merge == 'left_only'")
merged_left

# These are the counties with county_fips that we weren't able to find. They are not in the US zips csv, will have to find them
# elsewhere"
```

```python
merged_missing.sort_values(by='County_FIPS')
```

```python
# drop the missing counties from eeoc and replace them with their new fips
```

```python
print(f"EEOC should have {len(eeo)} rows.")
print(f"There are {len(string_fips)} rows without county codes ({string_fips['County_FIPS'].nunique()} counties).")
```

```python
mask = eeo.isin(string_fips)
eeo_without_missing_fips = eeo[~mask.all(axis=1)]
eeo_without_missing_fips

print(f"There are {len(eeo_without_missing_fips)} rows with county codes.")
print(f"We removed {len(eeo) - len(eeo_without_missing_fips)} rows without county codes.")
```

```python
# concat eeo_without_missing_fips and merged_both
```

```python
eeo_without_missing_fips
```

```python
columns = eeo.columns.to_list()
```

```python
merged_both_clean = merged_both.drop(['_merge'], axis='columns')
merged_both_clean
```

```python
columns = ['Year',
'County_FIPS',
 'County',
 'State',
 'Race',
 'Gender',
 'Industry',
 'Job',
 'Number of Employee'
]
```

```python
merged_both_clean = merged_both_clean[columns]
merged_both_clean
```

```python
merged_both_clean['County_FIPS'] = merged_both_clean['County_FIPS'].astype('int').astype('object')
merged_both_clean
```

```python
print(f"We fixed {merged_both_clean['County_FIPS'].nunique()} counties.")
print(f"There are {139 - 87} counties left to fix.")
```

```python
eeo_without_missing_fips
```

```python
eeo_new = pd.concat([eeo_without_missing_fips, merged_both_clean])
print(f"There are {len(eeo_new)} rows in the new EEOC")
print(f"There are {len(eeo) - len(eeo_new)} rows still without county_fips ({139-87} counties)")
```

```python
eeo_new.to_csv("EEOC_with_fips.csv", index=False)
```

```python
merged_left.to_csv('address_fips.csv', index=False)
```

```python
# Goal B: Fix the remaining missing FIPS
```

```python
merged_left_clean = merged_left.drop(['County_FIPS', '_merge'], axis='columns')
merged_left_clean
```

```python
merged_left_clean['county_state'] = merged_left_clean[['County', 'State']].apply((lambda x: x[0] +", " + x[1]), axis=1)
```

```python
print(f"We need to address {merged_left_clean['county_state'].nunique()} counties")
```

```python
missing_left_counties = merged_left_clean.drop_duplicates('county_state', keep='first')
missing_left_counties.head()
```

```python
missing_left_counties = missing_left_counties[['County', 'State']]
missing_left_counties.head()
```

```python
missing_left_counties.to_csv("remaining_missing_fips_raw.csv", index=False)
```

```python
last_missing_counties = pd.read_csv('remaining_missing_fips.csv')
last_missing_counties['county_state'] = last_missing_counties[['County', 'State']].apply((lambda x: x[0] + ", " + x[1]), axis = 1)
last_missing_counties = last_missing_counties.drop(['County', 'State'], axis="columns")


last_missing_counties.head()
```

```python
merged_left_clean.head()
```

```python
# merge last_missing_counties and merged_left_clean

finished_left_df = merged_left_clean.merge(right=last_missing_counties, on='county_state')
finished_left_df
```

```python
eeo_complete = pd.concat([eeo_new, finished_left_df])
eeo_complete = eeo_complete.drop(['county_state'], axis='columns')
```

```python
eeo_complete
```

```python
eeo_complete['County_FIPS'] = eeo_complete['County_FIPS'].astype('int').astype('object')
```

```python
eeo_complete.sort_values(by='County_FIPS')
```

```python
eeo_complete.to_csv("eeo_complete_v5.csv", index=False)
```
